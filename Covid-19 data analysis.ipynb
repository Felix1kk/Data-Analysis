## converting the date column to datetime datatype

adding new columns for day , month and year 

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.insert(4,'day',df['date'].dt.day_name())
```

```python
df.insert(6,'Year',pd.DatetimeIndex(df['date']).year)
```

## Locations with the highest death rate in the year 2020

```python

a=df.continent
a
```

```python
# about to perform a filter operation to remove these regions below from the dataframe

b=['World','High income','Upper middle income','European Union','Lower middle income']
```

```python
# Removing continents and other regions which are not country names from the dataframe
# filtering the location to get only countries from the dataframe and remove ambigious regions 

filtered=filt[(~filt['location'].isin(a)) & (~filt['location'].isin(b))]
locat=filtered.groupby('location')[['total_deaths',
                                   'new_deaths','new_deaths_smoothed']].sum().sort_values('total_deaths',ascending=False).head(20)
```

```python
locat
```

```python
plt.figure(figsize=(10,8))
sns.barplot(y=locat.index,x=locat.total_deaths)
plt.title('Top 20 locations with the highest death rate in the year 2020',fontsize=20)
plt.xlabel('total_deaths',fontsize=20)
plt.ylabel('location',fontsize=20)
```

 United states of America had the highest death rate in the year 2020

```python
avg_death_day=filtered.groupby('day')['total_deaths'].mean()
```

```python
plt.figure(figsize=(10,8))
plt.title('average death rate per day in the year 2020',fontsize=20)
sns.lineplot(x=avg_death_day.index, y=avg_death_day)
```

```python
plt.figure(figsize=(10,8))
avg_cases_day=filtered.groupby('day')['total_cases'].mean()
avg_cases_day
sns.lineplot(x=avg_cases_day.index,y=avg_cases_day)
```

```python
plt.figure(figsize=(12,6))
plt.title('average death rate per month in the year 2020',fontsize=20)
avg_death_month=filtered.groupby('month')['total_deaths'].mean()
sns.lineplot(x=avg_death_month.index, y=avg_death_month)
```

#### inference
1. December had the highest average death rate in the year 2020

```python
plt.figure(figsize=(10,6))
plt.title('average death rate per continent in the year 2020',fontsize=20)
avg_death_continent=filt.groupby('continent')['total_deaths'].mean()
sns.lineplot(x=avg_death_continent.index, y=avg_death_continent)
```

#### inference
1. south America had the highest average death rate in the year 2020
