```python
# for getting the month year and month values
df['month_year']=df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df.head()
```

```python
df_trend=df.groupby('month_year').sum()['sales'].reset_index()
```

```python
df_trend
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x=df_trend.month_year, y=df_trend.sales)
plt.xticks(rotation=79)
plt.show()
```

```python
# what are the top 10 product by sales
```

```python
most_sold=df.groupby('product_name').sum()['sales'].reset_index().sort_values(by='sales',ascending=False).head(10)
```

```python
most_sold
```

```python

plt.figure(figsize=(15,6))
sns.barplot(x=most_sold.product_name, y=most_sold.sales)
plt.xticks(rotation=79)
plt.show()
```

```python
cate=df.groupby('category').sum().sort_values('profit',ascending=False)
cate_av=df.groupby('category').mean().sort_values('profit',ascending=False)

sns.barplot( cate.index,cate.profit)
plt.title('category which made more profit',fontsize=20)
plt.show()
```
